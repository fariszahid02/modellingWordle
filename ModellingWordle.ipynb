* Guess the WORDLE in 6 tries.
* Each guess must be a valid 5 letter word. 
* The color of the tiles will change to show how close your guess was to the word.
* A green letter shows it is in the correct position.
* A yellow letter shows it is in the word but in the wrong position.
* A grey letter shows that it is not in the word in any position.

## Guess

We can now introduce a `Guess` type, which represents a guess after it is processed by the game (i.e. one line of the output). We use the `dataclass' decorator and properties to simplify definition. This includes the word that was guessed, and the clues.

1. There are two invariants to add to the `setter` functions; what are these? Complete the assertions and include an appropriate message for when they fail. These should restrict the values that `word` and `clues` can take. 
2. Create some test values for `Guess` to demonstrate the invariants, i.e.  `g1 = Guess("HELLO", [Clue.YELLOW, ...])`.

```python
@dataclass
class Guess:
    """
    A class to represent a guess in Wordle, which is a Word 
    plus a clue for each letter.
    """
    word: Word
    clues: List[Clue]

    @property
    def word(self) -> Word:
        return self._word

    @word.setter
    def word(self, word: Word):
        # invariant
        assert len(word) == WORD_LENGTH, \
            f"Invariant violated: len(Guess.word) <> {WORD_LENGTH}"  
        self._word = word

    @property
    def clues(self) -> List[Clue]:
        return self._clues

    @clues.setter
    def clues(self, clues: Clue):
        # invariant
        assert len(clues) == WORD_LENGTH, \
            f"Invariant violated: len(Guess.clues) <> {WORD_LENGTH}"              
        self._clues = clues

    def __repr__(self):
        """
        Custom representation for pretty printing.
        """
        cluestr = [str(self.word[i]) + ": " + \
            self.clues[i].name for i in range(WORD_LENGTH)]
        return f"{self.word}: {cluestr}"       

g1 = Guess("HELLO", [Clue.YELLOW, Clue.YELLOW, Clue.GREY, Clue.GREEN, Clue.YELLOW])
#g2 = Guess("HELL", [Clue.YELLOW, Clue.YELLOW, Clue.GREY, Clue.GREEN, Clue.YELLOW])
#g3 = Guess("HELLO", [Clue.YELLOW, Clue.YELLOW, Clue.GREY, Clue.GREEN]) 
```

## check_letter 

This function computes the clue (colour) for a given letter, given a word and the index of the letter in the word. We will start with a naïve implementation:

* If it's in the right place, return green (i.e. `Clue.GREEN`)
* If it’s not there, return grey
* Otherwise, return yellow.

```python
def check_letter(letter: str, index: int, word: Word) -> Clue:
    """
    Given a letter and an index, computes the colour of the blue
    based on the word.
    """
    # pre-condition
    assert 0 <= index < WORD_LENGTH and \
           len(word) == WORD_LENGTH, "pre-check_letter failed."
           
    if word[index] == letter: return Clue.GREEN  
    elif letter not in word: return Clue.GREY
    else: return Clue.YELLOW

print(check_letter("S", 0, "STOUT"))
print(check_letter("S", 3, "STOUT"))
print(check_letter("Z", 0, "STOUT"))
```

## Check Guess

Given a word and the correct answer, this function computes the list of clues. This can be achieved by declaring a local variable, and for each letter in the guess, use `check_letter`; we can use `enumerate(guess)` to get the index.

```python
def check_guess(word: Word, guess: Word) -> List[Clue]:
    """
    Given the answer and a guess, compute the list of 
    clues corresponding to each letter.
    """
    # pre-condition
    assert len(word) == WORD_LENGTH and \
           len(guess) == WORD_LENGTH, "pre-check_guess failed"
           
    clues = []
    for i,letter in enumerate(guess):
        clue = check_letter(guess[i], i ,word)
        clues.append(clue)
    return clues

print(check_guess("STAND", "STOUT"))
```

# Game State and Top-level Functionality

Here we define a `Game` class; the state is the current wordle (`answer`), the guesses made so far, and the state of the game (playing, won, or lost). Functionlity is included to print a message to the player about the game, and to rest the game when it is over.

The top-level function is `make_guesss` which takes a single word as a parameter.
```python
class Game:
    answer: Word
    guesses: List[Guess]
    gstate: Gamestate

    def __init__(self, answer="STOUT"):
        """
        Constructor for game.
        """
        self.answer = answer
        self.guesses = []
        self.gstate = Gamestate.PLAYING

    def make_guess(self, word: Word):
        """
        Make a guess at the wordle.
        """
        # make guesses uppercase
        word = word.upper() 

        # pre-condition
        assert self.gstate == Gamestate.PLAYING and \
               len(self.guesses) < MAX_GUESSES and \
               word in WORDS, "pre-guess failed."
               
        self.guesses.append(Guess(word, check_guess(self.answer, word)))
        if word == self.answer: self.gstate = Gamestate.WON
        elif len(self.guesses) == MAX_GUESSES: self.gstate = Gamestate.LOST

    def print_state(self):
        """
        Prints a message to the user based on the current state of the game.
        """
        for guess in self.guesses: print(guess)
        if self.gstate == Gamestate.WON: print(f"You won! You took {len(self.guesses)} guesses.")
        elif self.gstate == Gamestate.LOST: print(f"You lost! The answer was: {self.answer}.")
        else: print(f"Guess the wordle, you have {MAX_GUESSES - len(self.guesses)} guesses remaining.")
        
    def game_over(self) -> bool:
        """
        Yields true if the game is over (won or lost), false otherwise.
        """
        return self.gstate == Gamestate.WON or self.gstate == Gamestate.LOST 

    def reset(self):
        """
        Reset the game by picking a new word, clearing the guess, and
        setting the state back to playing.
        """
        # pre-condition
        assert self.game_over(), "Cannot reset, game in play"
        self.answer = random.choice(WORDS)
        self.guesses = []
        self.gstate = Gamestate.PLAYING
   
game = Game()
game.print_state()
game.make_guess("stand")
game.print_state()
```

## Improving 'check_guess' Function to Handle Repeated Letters

In this section, I've improved the `check_guess` function to resolve issues related to repeated letters and inconsistencies in providing the correct clues. This approach involves directly computing clues within `check_guess` rather than relying on `check_letter` function.

```python
def check_guess(word: Word, guess: Word) -> List[Clue]:
    """
    Guess and answer must be the same word length (5 letters)
    """
    assert len(word) == WORD_LENGTH and len(guess) == WORD_LENGTH
    
    clues = [Clue.GREY] * WORD_LENGTH
    word_counter = {}
    
    for i, letter in enumerate(word):
        if guess[i] != letter:
            word_counter[letter] = word_counter.get(letter, 0) + 1

    # mark clues green
    for i, letter in enumerate(guess):
        if word[i] == letter:
            clues[i] = Clue.GREEN
    
    # mark clues yellow if applicable
    for i, letter in enumerate(guess):
        if clues[i] == Clue.GREY and letter in word_counter and word_counter[letter] > 0:
            clues[i] = Clue.YELLOW
            word_counter[letter] -= 1

    return clues
```

## Hard Mode Implementation

In this section, I implemented a Hard Mode for my Wordle project. In Hard Mode, players must use any revealed hints in their future guesses. This rule adds an extra layer of challenge by limiting the options for what words users can guess next.


```python
class Game:
    answer: Word
    guesses: List[Guess]
    gstate: Gamestate
    
    def __init__(self, answer="STOUT"):
        """
        Constructor for game.
        """
        self.answer = answer
        self.guesses = []
        self.gstate = Gamestate.PLAYING

    def make_guess(self, word: Word):
        """
        Make a guess at the wordle.
        """
        # make guesses uppercase
        word = word.upper() 

        # pre-condition
        assert self.gstate == Gamestate.PLAYING and \
               len(self.guesses) < MAX_GUESSES and \
               word in WORDS, "pre-guess failed."
               
        self.guesses.append(Guess(word, check_guess(self.answer, word)))
        if word == self.answer: self.gstate = Gamestate.WON
        elif len(self.guesses) == MAX_GUESSES: self.gstate = Gamestate.LOST
        
    def hard_guess(self, word: Word):
        """
        Make a guess at the wordle. This method adheres to the rule of "Hard Mode".
        """
        # make guesses to uppercase
        word = word.upper()

        # pre-condition
        assert self.gstate == Gamestate.PLAYING and \
               len(self.guesses) < MAX_GUESSES and \
               word in WORDS, "Pre-guess failed."

        # pre-condition (hard mode)
        for previous_guess in self.guesses:
            for i, (g_letter, g_clue) in enumerate(zip(previous_guess.word, previous_guess.clues)):
                if g_clue == Clue.GREEN and word[i] != g_letter:
                    raise AssertionError("Guess does not follow the clues provided by GREEN hints.")
                elif g_clue == Clue.YELLOW and g_letter not in word:
                    raise AssertionError("Guess does not follow the clues provided by YELLOW hints.")

        self.guesses.append(Guess(word, check_guess(self.answer, word)))
        if word == self.answer: self.gstate = Gamestate.WON
        elif len(self.guesses) == MAX_GUESSES: self.gstate = Gamestate.LOST

    def print_state(self):
        """
        Prints a message to the user based on the current state of the game.
        """
        for guess in self.guesses: print(guess)
        if self.gstate == Gamestate.WON: print(f"You won! You took {len(self.guesses)} guesses.")
        elif self.gstate == Gamestate.LOST: print(f"You lost! The answer was: {self.answer}.")
        else: print(f"Guess the wordle, you have {MAX_GUESSES - len(self.guesses)} guesses remaining.")
        
    def hint(self) -> str:
        """
        Provides a hint to users according to the current state of the game.      
        """
        #pre-condition
        assert self.gstate == Gamestate.PLAYING, "Hints are only available if the game is playing."
        assert len(self.guesses) < MAX_GUESSES, "Hint cannot be given once the maximum number of guesses is reached."
        
        guessed_letters = {letter for guess in self.guesses for letter in guess.word}
        for letter in set(self.answer):
            if letter not in guessed_letters:
                return letter
        return ""

    def game_over(self) -> bool:
        """
        Yields true if the game is over (won or lost), false otherwise.
        """
        return self.gstate == Gamestate.WON or self.gstate == Gamestate.LOST 

    def reset(self):
        """
        Reset the game by picking a new word, clearing the guess, and
        setting the state back to playing.
        """
        # pre-condition
        assert self.game_over(), "Cannot reset, game in play"
        self.answer = random.choice(WORDS)
        self.guesses = []
        self.gstate = Gamestate.PLAYING

game = Game()
```

## Extension - Implementing Additional Invariants

In this section, I've defined additional invariants to maintain consistency in the `game` state. These invariants concentrate on keeping a logical relationship among `gstate`, `guesses`, and the maximum number of allowed guesses (MAX_GUESSES) throughout the game. The application of these invariants is evident in the `make_guess` and `hard_guess` methods.

```python
class Game:
    answer: Word
    guesses: List[Guess]
    gstate: Gamestate
    
    def __init__(self, answer="STOUT"):
        """
        Constructor for game.
        """
        self.answer = answer
        self.guesses = []
        self.gstate = Gamestate.PLAYING
    
    @property
    def guesses(self):
        return self._guesses

    @guesses.setter
    def guesses(self, value):
        # Ensure the number of guesses does not exceed the allowed maximum.
        assert len(value) <= MAX_GUESSES, "Invariant violated: More guesses than allowed."
        self._guesses = value
    
    def make_guess(self, word: Word):
        """
        Make a guess at the wordle.
        """
        # make guesses uppercase
        word = word.upper() 

        # pre-condition
        assert self.gstate == Gamestate.PLAYING and \
               len(self.guesses) < MAX_GUESSES and \
               word in WORDS, "pre-guess failed."
               
        self.guesses.append(Guess(word, check_guess(self.answer, word)))
        if word == self.answer: self.gstate = Gamestate.WON
        elif len(self.guesses) == MAX_GUESSES: self.gstate = Gamestate.LOST
        
        # invariant checks
        if self.gstate == Gamestate.WON:
            assert any(guess.word == self.answer for guess in self.guesses), \
                "Invariant violated: Game was incorrectly marked as won without a correct guess being made."
        if self.gstate == Gamestate.LOST:
            assert len(self.guesses) == MAX_GUESSES, \
                "Invariant violated: Game was incorrectly marked as lost before the maximum number of guesses was reached."
        
    def hard_guess(self, word: Word):
        """
        Make a guess at the wordle. This method adheres to the rule of "Hard Mode".
        """
        # make guesses to uppercase
        word = word.upper()

        # pre-condition
        assert self.gstate == Gamestate.PLAYING and \
               len(self.guesses) < MAX_GUESSES and \
               word in WORDS, "Pre-guess failed."

        # pre-condition (hard mode)
        for previous_guess in self.guesses:
            for i, (g_letter, g_clue) in enumerate(zip(previous_guess.word, previous_guess.clues)):
                if g_clue == Clue.GREEN and word[i] != g_letter:
                    raise AssertionError("Guess does not follow the clues provided by GREEN hints.")
                elif g_clue == Clue.YELLOW and g_letter not in word:
                    raise AssertionError("Guess does not follow the clues provided by YELLOW hints.")

        self.guesses.append(Guess(word, check_guess(self.answer, word)))
        if word == self.answer: self.gstate = Gamestate.WON
        elif len(self.guesses) == MAX_GUESSES: self.gstate = Gamestate.LOST
        
        # invariant checks
        if self.gstate == Gamestate.WON:
            assert any(guess.word == self.answer for guess in self.guesses), \
                "Invariant violated: Game was incorrectly marked as won without a correct guess being made."
        if self.gstate == Gamestate.LOST:
            assert len(self.guesses) == MAX_GUESSES, \
                "Invariant violated: Game marked as lost without reaching the max guesses."

    def print_state(self):
        """
        Prints a message to the user based on the current state of the game.
        """
        for guess in self.guesses: print(guess)
        if self.gstate == Gamestate.WON: print(f"You won! You took {len(self.guesses)} guesses.")
        elif self.gstate == Gamestate.LOST: print(f"You lost! The answer was: {self.answer}.")
        else: print(f"Guess the wordle, you have {MAX_GUESSES - len(self.guesses)} guesses remaining.")
        
    def hint(self) -> str:
        """
        Provides a hint to users according to the current state of the game.      
        """
        #pre-condition
        assert self.gstate == Gamestate.PLAYING, "Hints are only available if the game is playing."
        assert len(self.guesses) < MAX_GUESSES, "Hint cannot be given once the maximum number of guesses is reached."
        
        guessed_letters = {letter for guess in self.guesses for letter in guess.word}
        for letter in set(self.answer):
            if letter not in guessed_letters:
                return letter
        return ""

    def game_over(self) -> bool:
        """
        Yields true if the game is over (won or lost), false otherwise.
        """
        return self.gstate == Gamestate.WON or self.gstate == Gamestate.LOST 

    def reset(self):
        """
        Reset the game by picking a new word, clearing the guess, and
        setting the state back to playing.
        """
        # pre-condition
        assert self.game_over(), "Cannot reset, game in play"
        self.answer = random.choice(WORDS)
        self.guesses = []
        self.gstate = Gamestate.PLAYING

game = Game()
```

```python
class Game:
    answer: Word
    guesses: List[Guess]
    gstate: Gamestate
    
    def __init__(self, answer="STOUT"):
        """
        Constructor for game.
        """
        self.answer = answer
        self.guesses = []
        self.gstate = Gamestate.PLAYING
    
    @property
    def guesses(self):
        return self._guesses

    @guesses.setter
    def guesses(self, value):
        # Ensure the number of guesses does not exceed the allowed maximum.
        assert len(value) <= MAX_GUESSES, "Invariant violated: More guesses than allowed."
        self._guesses = value
    
    def make_guess(self, word: Word):
        """
        Make a guess at the wordle.
        """
        # make guesses uppercase
        word = word.upper() 

        # pre-condition
        assert self.gstate == Gamestate.PLAYING and \
               len(self.guesses) < MAX_GUESSES and \
               word in WORDS, "pre-guess failed."
               
        self.guesses.append(Guess(word, check_guess(self.answer, word)))
        if word == self.answer: self.gstate = Gamestate.WON
        elif len(self.guesses) == MAX_GUESSES: self.gstate = Gamestate.LOST
        
        # invariant checks
        if self.gstate == Gamestate.WON:
            assert any(guess.word == self.answer for guess in self.guesses), \
                "Invariant violated: Game was incorrectly marked as won without a correct guess being made."
        if self.gstate == Gamestate.LOST:
            assert len(self.guesses) == MAX_GUESSES, \
                "Invariant violated: Game was incorrectly marked as lost before the maximum number of guesses was reached."
        
    def hard_guess(self, word: Word):
        """
        Make a guess at the wordle. This method adheres to the rule of "Hard Mode".
        """
        # make guesses to uppercase
        word = word.upper()

        # pre-condition
        assert self.gstate == Gamestate.PLAYING and \
               len(self.guesses) < MAX_GUESSES and \
               word in WORDS, "Pre-guess failed."

        # pre-condition (hard mode)
        for previous_guess in self.guesses:
            for i, (g_letter, g_clue) in enumerate(zip(previous_guess.word, previous_guess.clues)):
                if g_clue == Clue.GREEN and word[i] != g_letter:
                    raise AssertionError("Guess does not follow the clues provided by GREEN hints.")
                elif g_clue == Clue.YELLOW and g_letter not in word:
                    raise AssertionError("Guess does not follow the clues provided by YELLOW hints.")

        self.guesses.append(Guess(word, check_guess(self.answer, word)))
        if word == self.answer: self.gstate = Gamestate.WON
        elif len(self.guesses) == MAX_GUESSES: self.gstate = Gamestate.LOST
        
        # invariant checks
        if self.gstate == Gamestate.WON:
            assert any(guess.word == self.answer for guess in self.guesses), \
                "Invariant violated: Game was incorrectly marked as won without a correct guess being made."
        if self.gstate == Gamestate.LOST:
            assert len(self.guesses) == MAX_GUESSES, \
                "Invariant violated: Game marked as lost without reaching the max guesses."

    def print_state(self):
        print("\nCurrent game state:")
        for guess in self._guesses:
            wordle = "".join(
                f"+{letter}" if clue == Clue.GREEN else
                f"-{letter}" if clue == Clue.YELLOW else
                f" {letter}" for letter, clue in zip(guess.word, guess.clues)
            )
            print(wordle)
        print("\nRemaining guesses:", MAX_GUESSES - len(self._guesses))

        if self.gstate == Gamestate.PLAYING:
            print("Keep guessing...")
        elif self.gstate == Gamestate.WON:
            print("Congratulations, you've won!")
        elif self.gstate == Gamestate.LOST:
            print("Game over. Better luck next time!")
            print("The answer was:", self.answer)
            
    def hint(self) -> str:
        """
        Provides a hint to users according to the current state of the game.      
        """
        #pre-condition
        assert self.gstate == Gamestate.PLAYING, "Hints are only available if the game is playing."
        assert len(self.guesses) < MAX_GUESSES, "Hint cannot be given once the maximum number of guesses is reached."
        
        guessed_letters = {letter for guess in self.guesses for letter in guess.word}
        for letter in set(self.answer):
            if letter not in guessed_letters:
                return letter
        return ""

    def game_over(self) -> bool:
        """
        Yields true if the game is over (won or lost), false otherwise.
        """
        return self.gstate == Gamestate.WON or self.gstate == Gamestate.LOST 

    def reset(self):
        """
        Reset the game by picking a new word, clearing the guess, and
        setting the state back to playing.
        """
        # pre-condition
        assert self.game_over(), "Cannot reset, game in play"
        self.answer = random.choice(WORDS)
        self.guesses = []
        self.gstate = Gamestate.PLAYING

game = Game()
```
